```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dropout,Dense
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('cleantraining.csv')
UnrankedGames = pd.read_csv('cleantesting.csv')
UnrankedGames =UnrankedGames.reindex(columns=data.columns)
UnrankedX = UnrankedGames.drop(['id','bayesaverage','name','descriptions','mechanics','categories'], axis=1)
y=data['bayesaverage']
X=data.drop(['id','bayesaverage','name','descriptions','mechanics','categories'], axis=1)
```

```python
y=y/10
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=9874) #Just for testing, full mod will use all of X
```

```python
model = Sequential([
    Dense(units=16, input_dim=556, activation='sigmoid'),
    Dense(units=8, activation='sigmoid'),
    Dense(units=1, activation='sigmoid')
])


optimizer = Adam(learning_rate=0.0005)
model.compile(optimizer=optimizer,loss='MeanSquaredError', metrics=['accuracy'])
model.fit(X, y, epochs= 600, batch_size=50)
```

```python
y_hat = model.predict(X_test)
np.sqrt(mean_squared_error(y_test,y_hat))
```

```python
unrankedpreds = model.predict(UnrankedX)
```

```python
unrankedpreds10 = unrankedpreds*10
```

```python
unrankedpreds10=unrankedpreds10.flatten()
```

```python
predDF = pd.DataFrame({'Name': UnrankedGames['name'],
                       'PredDNN': unrankedpreds10,
                       'OldRating': UnrankedGames['bayesaverage']})
```

```python
predDF.to_csv('DeeplearningPredictions.csv',index=False)
```
